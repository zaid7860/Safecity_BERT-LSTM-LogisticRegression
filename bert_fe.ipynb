```
train_df =  pd.read_csv('/content/drive/MyDrive/data/train 2.csv',nrows=5000,names=['Description','Commenting','Staring','Touching'])
test_df=pd.read_csv('/content/drive/MyDrive/data/test 2.csv',nrows=1500,names=['Description','Commenting','Staring','Touching'])
```

```
import numpy as np
from transformers import AutoTokenizer, AutoModel, pipeline


model = AutoModel.from_pretrained('distilbert-base-uncased')
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')
    
nlp = pipeline('feature-extraction', model=model, tokenizer=tokenizer,)
def convertion(string):
    
    output = nlp(string)
    output = np.array(output)
    output = output[:,0,:]
    #output = np.array(output)
    #output=output[:,0,:]

 
    return np.array(output)


import numpy as np


#model = AutoModel.from_pretrained('distilbert-base-uncased')
#tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')
#nlp = pipeline('feature-extraction', model=model, tokenizer=tokenizer
```

```
from tqdm import tqdm
sent =[]
train_vec=[]
for i in tqdm((train_df['Description'].values)):
    text = i[:512]
    train_vec.append(convertion(text))
```

```
train_vec=np.array(train_vec)

train_vec.shape
```

```
train_vec=train_vec.reshape(5000,768)
```

```
train_vec.shape
```

```
from tqdm import tqdm

test_vec=[]
for i in tqdm(test_df['Description']):
    text = i[:512]
    test_vec.append(convertion(text))
```

```
test_vec=np.array(test_vec)

test_vec.shape
```

```
test_vec=test_vec.reshape(1500,768)
```

```
test_vec.shape
```

```
from sklearn.linear_model import LogisticRegression
```

```
#x_train=train_df['Vector']
y_train=train_df['Commenting']
##y_train=train_df['Ogling/Facial Expressions/Staring']


#x_test=test_df['Vector']
y_test=test_df['Commenting']
```

```
features=train_vec
```

```
lr_clf = LogisticRegression()
lr_clf.fit(features, y_train)
```

```
y_pred = lr_clf.predict(test_vec)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(lr_clf.score(test_vec, y_test)))
```

```
lr_clf.score(test_vec, y_test)
```

How Good is this model as compared to that of Random Model

```
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import cross_val_score

clf = DummyClassifier()

scores = cross_val_score(clf, features, y_train)
print("Dummy classifier score: %0.3f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
